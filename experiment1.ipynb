```python
def is_palindrome(number):
    # Convert the number to a string for easier comparison
    num_str = str(number)

    # Compare the string with its reverse
    if num_str == num_str[::-1]:
        print(f"The number {number} is a palindrome.")
    else:
        print(f"The number {number} is not a palindrome.")

# Test cases
is_palindrome(121)
is_palindrome(12321)
is_palindrome(12345)
is_palindrome(0)
is_palindrome(7)
is_palindrome(1001)
is_palindrome(-121) # Note: Negative numbers are generally not considered palindromes in this context
```

```python
def check_odd_even(number):
    if isinstance(number, int):
        if number % 2 == 0:
            print(f"The number {number} is Even.")
        else:
            print(f"The number {number} is Odd.")
    else:
        print(f"Invalid input: {number}. Please enter an integer.")

# Test cases
check_odd_even(4)
check_odd_even(7)
check_odd_even(0)
check_odd_even(-3)
check_odd_even(10.5)
check_odd_even("hello")
```

```python
def display_multiplication_table(number):
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):  # Table from 1 to 10
        print(f"{number} x {i} = {number * i}")

# Test with a number
num_to_display = 7
display_multiplication_table(num_to_display)

print("\n") # Add a newline for better readability between outputs

num_to_display = 12
display_multiplication_table(num_to_display)
```

```python
def factorial(n):
    if not isinstance(n, int) or n < 0:
        return "Input must be a non-negative integer."
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Test cases
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = 7
print(f"The factorial of {number} is {factorial(number)}")

number = -3
print(f"The factorial of {number} is {factorial(number)}")

number = 3.5
print(f"The factorial of {number} is {factorial(number)}")
```
